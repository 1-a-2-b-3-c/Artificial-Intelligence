```python
import random
import copy
from collections import deque
```

```python
class Node:
    def __init__(self, state, parent, direction, x, y):
        self.state = state
        self.parent = parent
        self.direction = direction
        self.x = x
        self.y = y

    def print_state(self):
        for row in self.state:
            print(row)
        print()
```

```python
def print_father_to_child(list_node):
    father = list_node[0].parent

    print("Đường đi từ cha đến con:", end=" ")
    for i in range(len(list_node)):
        match list_node[i].direction:
            case "up":
                print("⬆️", end="  ")
            case "down":
                print("⬇️", end="  ")
            case "left":
                print("⬅️", end="  ")
            case "right":
                print("➡️", end="  ")
    print()

    for i in range(3):
        print(father.state[i], end="")
        if i == 1:
            print("   ->   ", end="")
        else:
            print("        ", end="")
        for j in range(len(list_node)):
            if j != 0:
                print("  |  ", end="")
            print(list_node[j].state[i], end="")
        print()
    print()
```

```python
def check_solvable(numbers):
    count = 0
    for i in range(9):
        for j in range(i + 1, 9):
            if numbers[i] != 0 and numbers[j] != 0 and numbers[i] > numbers[j]:
                count += 1
    return count % 2 == 0
```

```python
def create_input():
    goal = [[1, 2, 3], [4, 5, 6], [7, 8, 0]]
    frontier = deque()
    readed = set()

    numbers = [0, 1, 2, 3, 4, 5, 6, 7, 8]
    while True:
        random.shuffle(numbers)
        if check_solvable(numbers):
            break
    state = [
            numbers[0:3],
            numbers[3:6],
            numbers[6:9]  
        ]

    x, y = 0, 0
    for i in range(3):
        for j in range(3):
            if state[i][j] == 0:
                x, y = i, j
    state = Node(state, None, None, x, y)

    return state, goal, frontier, readed
```

```python
def move_possible(x, y):
    moves = []
    if x > 0:
        moves.append('up')
    if x < 2:
        moves.append('down')
    if y > 0:
        moves.append('left')
    if y < 2:
        moves.append('right')
    return moves
```

```python
def print_solution(node):
    solution = []
    while node:
        solution.append(node.state)
        node = node.parent
    solution.reverse()

    print(f"Số bước: {len(solution) - 1}")
    for state in solution:
        for row in state:
            print(row)
        print()
```

```python
def action(x, y, state, direction):
    new_state = copy.deepcopy(state)
    match direction:
        case 'up':
            new_state[x][y], new_state[x-1][y] = new_state[x-1][y], new_state[x][y]
            x -= 1
        case 'down':
            new_state[x][y], new_state[x+1][y] = new_state[x+1][y], new_state[x][y]
            x += 1
        case 'left':
            new_state[x][y], new_state[x][y-1] = new_state[x][y-1], new_state[x][y]
            y -= 1
        case 'right':
            new_state[x][y], new_state[x][y+1] = new_state[x][y+1], new_state[x][y]
            y += 1
    return new_state, x, y
```

```python
def bfs(node, goal, frontier, readed):
    frontier.append(node)
    readed.add(tuple(map(tuple, node.state)))

    print("Trạng thái ban đầu:")
    node.print_state()

    i = 0
    while True:
        i += 1

        if not frontier:
            print("Không còn trạng thái nào để duyệt.")
            break

        current_node = frontier.popleft()

        if current_node.state == goal:
            print("Đã tìm thấy giải pháp!")
            print_solution(current_node)
            return 
        
        list_child = []
        for move in move_possible(current_node.x, current_node.y):
            new_state, new_x, new_y = action(current_node.x, current_node.y, current_node.state, move)

            if tuple(map(tuple, new_state)) not in readed:
                new_node = Node(new_state, current_node, move, new_x, new_y)
                list_child.append(new_node)
                frontier.append(new_node)
                readed.add(tuple(map(tuple, new_state)))

        print(f"Duyệt qua trạng thái thứ {i}.")
        if current_node.parent == None:
            current_node.print_state()
        else:
            if list_child:
                print_father_to_child(list_child)
            else:    
                print("Không tồn tại đường đi nào hợp lệ.")  


    print(f"Đã duyệt qua {i} trạng thái.")
    print("Không tìm thấy giải pháp trong giới hạn duyệt.")
   
```

```python
node, goal, frontier, readed = create_input()
bfs(node, goal, frontier, readed)
```
