```python
import random
import copy
```

```python
def create_input():
    goal = [[1, 2, 3], [4, 5, 6], [7, 8, 0]]
    numbers = [0, 1, 2, 3, 4, 5, 6, 7, 8]
    random.shuffle(numbers)
    state = [
        numbers[0:3],
        numbers[3:6],
        numbers[6:9]
    ]
    state_past= []
    x, y = 0, 0
    for i in range(3):
        for j in range(3):
            if state[i][j] == 0:
                x, y = i, j
    return goal, state, x, y, state_past
```

```python
def move_piece(move, x, y, state):
    match move:
        case "up":
            state[x][y], state[x-1][y] = state[x-1][y], state[x][y]
            x -= 1
        case "down":
            state[x][y], state[x+1][y] = state[x+1][y], state[x][y]
            x += 1
        case "left":
            state[x][y], state[x][y-1] = state[x][y-1], state[x][y]
            y -= 1
        case "right":
            state[x][y], state[x][y+1] = state[x][y+1], state[x][y]
            y += 1
    return x, y, state
```

```python
def can_move(x, y):
    moves = []
    if x > 0:
        moves.append("up")
    if x < 2:
        moves.append("down")
    if y > 0:
        moves.append("left")
    if y < 2:
        moves.append("right")
    return moves
```

```python
def rule_move(x, y, state, state_past):
    moves = can_move(x, y)
    for move in moves:
        new_x, new_y, new_state = move_piece(move, x, y, copy.deepcopy(state))
        if new_state in state_past:
            moves.remove(move)
    return moves
```

```python
def print_board(state):
    for row in state:
        print(row)
    print()
```

```python
def action(x, y, state, state_past, goal):
    print("Trạng thái ban đầu:")
    print_board(state)

    if state == goal:
        print("Đã đạt trạng thái mục tiêu ngay từ đầu!")
        return

    i = 0
    while i < 2000:
        i += 1

        moves = rule_move(x, y, state, state_past)
        print(f"Các nước đi có thể thực hiện lần {i}: {moves}")
        if moves == []:
            print("Không thể di chuyển nữa. Đã đạt trạng thái đã đi qua.")
            break

        move = random.choice(moves)
        x, y, state = move_piece(move, x, y, state)
        if state == goal:
            print(f"Đã đạt trạng thái mục tiêu sau {i} lần di chuyển!")
            print_board(state)
            break

        state_past.append(copy.deepcopy(state))
        print(f"Thực hiện: {move}")
        print_board(state)

    print("Sau 2000 lần di chuyển, chưa đạt trạng thái mục tiêu.")
    print("Kết thúc chương trình.")
    print()
```

```python
goal, state, x, y, state_past = create_input()
action(x, y, state, state_past, goal)
```
