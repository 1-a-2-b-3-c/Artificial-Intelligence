```python
import random
import copy
```

```python
def create_input():
    m = random.randint(3, 7)
    n = random.randint(3, 7)
    state = [[random.choice([-1, 0, 1]) for _ in range(n)] for _ in range(m)]

    goal = [[0] * n for _ in range(m)]
    for i in range(m):
        for j in range(n):
            if state[i][j] == 1:
                goal[i][j] = 0

    while True:
        x,y = random.randint(0, m-1), random.randint(0, n-1)
        if state[x][y] != -1:
            break

    return m, n, state, goal, x, y
```

```python
def print_state_no_robot(m, n, state):
    for i in range(m):
        print("[ ", end="")
        for j in range(n):
            if state[i][j] == 0:
                print("0", end=" ")
            elif state[i][j] == 1:
                print("*", end=" ")
            else:
                print("■", end=" ")
        print(" ]")
    print()
```

```python
def print_state_with_robot(m, n, state, x, y):
    for i in range(m):
        print("[ ", end="")
        for j in range(n):
            if i == x and j == y:
                print("◉", end=" ")
            elif state[i][j] == 0:
                print("0", end=" ")
            elif state[i][j] == 1:
                print("*", end=" ")
            else:
                print("■", end=" ")
        print(" ]")
    print()
```

```python
def move_piece(move, x, y):
    match move:
        case 'up':
            x -= 1
        case 'down':
            x += 1
        case 'left':
            y -= 1
        case 'right':
            y += 1
        case _:
            return -1, -1 
    return x, y
```

```python
def can_move(x, y, m, n, state):
    moves = []
    if x > 0 and state[x-1][y] != -1:
        moves.append('up')
    if x < m - 1 and state[x+1][y] != -1:
        moves.append('down')
    if y > 0 and state[x][y-1] != -1:
        moves.append('left')
    if y < n - 1 and state[x][y+1] != -1:
        moves.append('right')
    return moves
```

```python
def rule_move(x, y, state, state_past):
    moves = can_move(x, y, m, n, state)
    valid_moves = []

    for move in moves:
        new_x, new_y = move_piece(move, x, y)
        if [state, (new_x, new_y)] not in state_past:
            valid_moves.append(move)

    return valid_moves
```

```python
def action(x, y, state, state_past, goal):
    print("Trạng thái ban đầu:")
    print_state_no_robot(m, n, state)

    print("Đặt robot tại vị trí (", x, ",", y, ")")
    if state[x][y] == 1:
        print("Robot đã ở trên một ô bẩn. Vệ sinh ô này.")
        state[x][y] = 0
    print_state_with_robot(m, n, state, x, y)

    state[x][y] = 0
    state_past.append([copy.deepcopy(state), (x, y)])
    if state == goal:
            print("Đã đạt trạng thái mục tiêu!")
    
    i = 0
    while i < 10000:
        
        i += 1

        moves = rule_move(x, y, state, state_past)
        if moves == []:
            print("Không còn nước đi hợp lệ nào nữa. Kết thúc.")
            break
        print(f"Các nước đi có thể thực hiện lần {i}: {moves}")

        move = random.choice(moves)
        print(f"Chọn nước đi: {move}")
        x, y = move_piece(move, x, y)

        if state[x][y] == 1:
            print("Robot đã ở trên một ô bẩn. Vệ sinh ô này.")
            state[x][y] = 0
        else:
            print("Robot đã ở trên một ô sạch.")

        print_state_with_robot(m, n, state, x, y)
        state_past.append([copy.deepcopy(state), (x, y)])

        if state == goal:
            print("Đã đạt trạng thái mục tiêu!")
            break

    print("Sau 10000 lần di chuyển, chưa đạt trạng thái mục tiêu.")
    print("Kết thúc chương trình.")
    print()
```

```python
m, n, state, goal, x, y = create_input()
state_past = []
action(x, y, state, state_past, goal)
```
