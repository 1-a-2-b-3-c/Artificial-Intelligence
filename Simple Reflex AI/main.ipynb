```python
import random
import copy

goal = [[1,2,3],[4,5,6],[7,8,0]]
numbers = [0,1,2,3,4,5,6,7,8]
random.shuffle(numbers)
board = [
    numbers[0:3],
    numbers[3:6],
    numbers[6:9]
]
boardList = []
x,y = 0,0

for i in range(3):
    for j in range(3):
        if board[i][j] == 0:
            x, y = i, j

def move_piece(move, x, y, board):
    match move:
        case "up":
            board[x][y], board[x-1][y] = board[x-1][y], board[x][y]
            x -= 1
        case "down":
            board[x][y], board[x+1][y] = board[x+1][y], board[x][y]
            x += 1
        case "left":
            board[x][y], board[x][y-1] = board[x][y-1], board[x][y]
            y -= 1
        case "right":
            board[x][y], board[x][y+1] = board[x][y+1], board[x][y]
            y += 1
    return x,y,board


def printState(board, i):
    print(f"State {i}:")
    for row in board:
        print(row)
    print()

def move():
    global x, y, board
    i = 1
    while True:
        boardList.append(copy.deepcopy(board))
        if x == 0 and y == 0:
            randomMove = random.choice(["down", "right"])
        elif x == 0 and y == 1:
            randomMove = random.choice(["down", "left", "right"])
        elif x == 0 and y == 2:
            randomMove = random.choice(["down", "left"])
        elif x == 1 and y == 0:
            randomMove = random.choice(["up", "down", "right"])
        elif x == 1 and y == 1:
            randomMove = random.choice(["up", "down", "left", "right"])
        elif x == 1 and y == 2:
            randomMove = random.choice(["up", "down", "left"])
        elif x == 2 and y == 0:
            randomMove = random.choice(["up", "right"])
        elif x == 2 and y == 1:
            randomMove = random.choice(["up", "left", "right"])
        elif x == 2 and y == 2:
            randomMove = random.choice(["up", "left"])
        printState(board,i)
        x,y,board = move_piece(randomMove, x, y, board)
        i += 1
        if board == goal:
            printState(board, i)
            print("Hoàn thành!")
            break
        if board in boardList:
            printState(board, i)
            print("Lặp lại trạng thái cũ")
            break
        if i > 100:
            printState(board, i)
            print("Không hoàn thành trong 100 nước")
            break

printState(board, 0)
move()
```
