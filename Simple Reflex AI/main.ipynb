PEAS của máy hút bụi
P: Hướng tối ưu để sàn nhà sạch trong khoảng thời gian ngắn nhất
E: Sàn nhà, bụi
A: Hàm để di chuyển, hàm để quét, hàm để hút bụi
S: Cảm biến bụi, cảm biến khoảng cách, cảm biến va chạm

```python
import random
```

```python
def interept_input():
    m,n = random.randint(3,6), random.randint(3,6)
    house =[]
    for i in range(n):
        row = []
        for j in range(m):
            row.append(random.choice([0,1,2]))
        house.append(row)
    goal = []

    for i in range(n):
        row = []

        for j in range(m):
            if house[i][j] == 2:
                row.append(2)
            else:
                row.append(0)
        goal.append(row)

    return m,n,house,goal
```

```python
def start_index(house):
    while True:
        x,y = random.randint(0,n-1), random.randint(0,m-1)

        if house[x][y] != 2:
            break
    return x,y
```

```python
def state(x, y):
    if house[x][y] == 1:
        return "Dirty"
    return "Clean"
```

```python
def rule_check(x, y, house):
    if house[x][y] == 1:
        house[x][y] = 0
        print(f"Vị trí đã dọn dẹp: ({y}, {x})")
```

```python
def move_piece(move, x, y):
    match move:
        case "up":
            x -= 1
        case "down":
            x += 1
        case "left":
            y -= 1
        case "right":
            y += 1
        case _:
            x,y = -1,-1

    return x,y
```

```python
def rule_move(x, y, m, n, house):
    moves = ["up", "down", "left", "right"]
    if x == 0:
        moves.remove("up")
    if x == n-1:
        moves.remove("down")
    if y == 0:
        moves.remove("left")
    if y == m-1:
        moves.remove("right")

    if x > 0 and house[x-1][y] == 2 and "up" in moves:
        moves.remove("up")
    if x < n-1 and house[x+1][y] == 2 and "down" in moves:
        moves.remove("down")
    if y > 0 and house[x][y-1] == 2 and "left" in moves:
        moves.remove("left")
    if y < m-1 and house[x][y+1] == 2 and "right" in moves:
        moves.remove("right")

    if len(moves) == 0:
        return -1, -1
        
    randomMove = random.choice(moves)
    return move_piece(randomMove, x, y)
```

```python
def check_move(x, y):
    if x == -1 or y == -1:
        return False
    return True
```

```python
def printHouseHaveRobot(house, x, y):
    for i in range(n):
        print("[", end="")
        for j in range(m):
            if i == x and j == y:
                print("◉", end=" ")
            else:
                print(house[i][j], end=" ")
        print("]")
    print()
```

```python
def printHouseNoRobot(house, x, y):
    for i in range(n):
        print("[", end="")
        for j in range(m):
            print(house[i][j], end=" ")
        print("]")
```

```python
def start(x,y,house):
    print(f"Vị trí ban đầu: ({y}, {x})")
    if state(x, y) == "Dirty":
        print(f"Vị trí ban đầu đã bẩn, dọn dẹp tại vị trí ({y}, {x})")
        rule_check(x, y, house)
    else:
        print(f"Vị trí ban đầu đã sạch, không cần dọn dẹp tại vị trí ({y}, {x})")
    printHouseHaveRobot(house, x, y)
```

```python
def action(x,y,house,goal):
    i = 0

    print("Sàn nhà ban đầu:")
    printHouseNoRobot(house, x, y)

    start(x,y,house)

    while True:
        i += 1
        x,y = rule_move(x, y, m, n, house)
        if check_move(x, y) == False:
            print(f"Di chuyển lần {i}: Không thể di chuyển ra ngoài, giữ nguyên vị trí ({y}, {x})")
            return 
            

        if state(x, y) == "Dirty":
            print(f"Di chuyển lần {i}: Dọn dẹp tại vị trí ({y}, {x})")
            rule_check(x, y, house)
        else:
            print(f"Di chuyển lần {i}: Vị trí ({y}, {x}) đã sạch")
        
        if house == goal:
            print("Hoàn thành!")
            break

        printHouseHaveRobot(house, x, y)

        if i == 100:
            print("Không hoàn thành sau 100 lần di chuyển")
            break
```

```python
m,n,house,goal = interept_input()
x,y = start_index(house)
action(x,y,house,goal)
```
